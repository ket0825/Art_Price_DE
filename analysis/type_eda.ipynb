```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("csv/picasso_Drawing.csv", encoding='utf-8')
df2 = pd.read_csv("csv/picasso_Gouache Painting.csv", encoding='utf-8')
df3 = pd.read_csv("csv/picasso_Oil Painting.csv", encoding='utf-8')
df4 = pd.read_csv("csv/picasso_Watercolor Painting.csv", encoding='utf-8')
df5 = pd.read_csv("csv/picasso_Print.csv", encoding='utf-8')
```

```python
df1 = df1.dropna(subset=['low_estimate_USD', 'high_estimate_USD', 'price_USD'])
df2 = df2.dropna(subset=['low_estimate_USD', 'high_estimate_USD', 'price_USD'])
df3 = df3.dropna(subset=['low_estimate_USD', 'high_estimate_USD', 'price_USD'])
df4 = df4.dropna(subset=['low_estimate_USD', 'high_estimate_USD', 'price_USD'])
df5 = df5.dropna(subset=['low_estimate_USD', 'high_estimate_USD', 'price_USD'])
```

```python
df_meta = [(df1, 'Drawing'), (df2, 'Gouache Painting'), (df3, 'Oil Painting'), (df4, 'Watercolor Painting'), (df5, 'Print')]
```

```python
for df, meta in df_meta:
    df['end_date'] = pd.to_datetime(df['end_date'])
    df['lot_year'] = df['end_date'].dt.year

    # filter
    # 상하위 5% 필터링
    filter_percent = 0.05
    lower_bound = df['price_USD'].quantile(filter_percent)
    upper_bound = df['price_USD'].quantile(1 - filter_percent)
    
    # 필터링된 데이터프레임 생성
    df_filtered = df[(df['price_USD'] >= lower_bound) & (df['price_USD'] <= upper_bound)].copy()
    
    # 필터링 전후 데이터 수 출력
    print(f"\n{meta} - 필터링 결과:")
    print(f"전체 데이터: {len(df)} 건")
    print(f"필터링 후: {len(df_filtered)} 건")
    print(f"제외된 하위 5% 기준가: ${lower_bound:,.2f}")
    print(f"제외된 상위 5% 기준가: ${upper_bound:,.2f}")

    groupby_count = df_filtered['lot_year'].value_counts().reset_index(name='count')

    lot_year_groupby = df_filtered.groupby('lot_year').agg({
        'low_estimate_USD': 'mean',
        'high_estimate_USD': 'mean',
        'price_USD': 'mean',
    }).reset_index()
    
    # 예상가 중앙값 계산
    lot_year_groupby['estimate_median'] = (lot_year_groupby['low_estimate_USD'] + lot_year_groupby['high_estimate_USD']) / 2
    # 실제가와 예상가 중앙값의 차이를 퍼센트로 계산
    lot_year_groupby['price_diff_pct'] = ((lot_year_groupby['price_USD'] - lot_year_groupby['estimate_median']) / lot_year_groupby['estimate_median']) * 100
    
    lot_year_groupby = pd.merge(lot_year_groupby, groupby_count, on='lot_year')
    
    # 그래프 생성
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 12), height_ratios=[3, 1, 1])
    
    # 첫 번째 플롯: 가격 범위와 실제가
    ax1.set_yscale('log')
    ax1.fill_between(lot_year_groupby['lot_year'],
                    lot_year_groupby['low_estimate_USD'],
                    lot_year_groupby['high_estimate_USD'],
                    alpha=0.2, color='blue', label='Estimate Range')
    ax1.plot(lot_year_groupby['lot_year'], lot_year_groupby['estimate_median'], 
             'b--', label='Estimate Median', alpha=0.7)
    ax1.plot(lot_year_groupby['lot_year'], lot_year_groupby['price_USD'], 
             'r-', label='Actual Price', linewidth=2)
    ax1.set_ylabel('Price (USD) - Log Scale')
    ax1.set_xlim(lot_year_groupby['lot_year'].min(), lot_year_groupby['lot_year'].max()+1)
    ax1.legend(loc='upper left')
    
    # 두 번째 플롯: 거래량
    ax2.bar(lot_year_groupby['lot_year'], lot_year_groupby['count'], 
            color='gray', alpha=0.5, label='Count')
    ax2.set_ylabel('Count')
    ax2.set_xlim(lot_year_groupby['lot_year'].min(), lot_year_groupby['lot_year'].max()+1)
    ax2.legend(loc='upper left')
    
    # 세 번째 플롯: 예상가 대비 실제가 차이 (퍼센트)
    bars = ax3.bar(lot_year_groupby['lot_year'], lot_year_groupby['price_diff_pct'], 
                   color=['red' if x < 0 else 'green' for x in lot_year_groupby['price_diff_pct']], 
                   alpha=0.5, label='Price Difference %')
    ax3.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    ax3.set_ylabel('Difference from Estimate (%)')
    ax3.set_xlim(lot_year_groupby['lot_year'].min(), lot_year_groupby['lot_year'].max()+1)
    ax3.set_xlabel('Year')
    
    # 평균 차이 퍼센트 계산 및 표시
    mean_diff = lot_year_groupby['price_diff_pct'].mean()
    ax3.text(0.02, 0.95, f'Average Difference: {mean_diff:.1f}%', 
             transform=ax3.transAxes, bbox=dict(facecolor='white', alpha=0.8))
    
    plt.suptitle(f'Picasso {meta} Price Trend and Estimate Comparison')
    plt.tight_layout()
    plt.show()
    
    # 연도별 차이 출력
    print(f"\n{meta} - 연도별 예상가 대비 실제가 차이:")
    for _, row in lot_year_groupby.iterrows():
        print(f"{row['lot_year']}: {row['price_diff_pct']:.1f}%")
    print(f"평균 차이: {mean_diff:.1f}%")
```

```python
for df, meta in df_meta:
    df['end_date'] = pd.to_datetime(df['end_date'])
    df['lot_year'] = df['end_date'].dt.year
    df = df[df['lot_year'] != 2025] # 2025년 데이터 제외
    
    # 상하위 5% 필터링
    filter_percent = 0.05
    lower_bound = df['price_USD'].quantile(filter_percent)
    upper_bound = df['price_USD'].quantile(1 - filter_percent)
    df_filtered = df[(df['price_USD'] >= lower_bound) & (df['price_USD'] <= upper_bound)].copy()

    print(f"제외된 하위 5% 기준가: ${lower_bound:,.2f}")
    print(f"제외된 상위 5% 기준가: ${upper_bound:,.2f}")
    
    groupby_count = df_filtered['lot_year'].value_counts().reset_index(name='count')

    lot_year_groupby = df_filtered.groupby('lot_year').agg({
        'low_estimate_USD': 'mean',
        'high_estimate_USD': 'mean',
        'price_USD': 'mean',
    }).reset_index()
    
    # 예상가 중앙값과 차이 계산
    lot_year_groupby['estimate_median'] = (lot_year_groupby['low_estimate_USD'] + lot_year_groupby['high_estimate_USD']) / 2
    lot_year_groupby['price_diff_pct'] = ((lot_year_groupby['price_USD'] - lot_year_groupby['estimate_median']) / lot_year_groupby['estimate_median']) * 100
    
    # 지수가중이동평균 계산 (span=3년으로 설정)
    lot_year_groupby['ewm_diff'] = lot_year_groupby['price_diff_pct'].ewm(span=3, adjust=False).mean()
    
    # 다음 년도 예측
    last_year = lot_year_groupby['lot_year'].max()
    last_ewm = lot_year_groupby['ewm_diff'].iloc[-1]
    
    # 예측 신뢰구간 계산 (최근 3년 변동성 기준)
    recent_std = lot_year_groupby['price_diff_pct'].tail(3).std()
    prediction_upper = last_ewm + recent_std
    prediction_lower = last_ewm - recent_std
    
    lot_year_groupby = pd.merge(lot_year_groupby, groupby_count, on='lot_year')
    
    # 그래프 생성
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 12), height_ratios=[3, 1, 1])
    
    # 첫 번째 플롯: 가격 범위와 실제가
    ax1.set_yscale('log')
    ax1.fill_between(lot_year_groupby['lot_year'],
                    lot_year_groupby['low_estimate_USD'],
                    lot_year_groupby['high_estimate_USD'],
                    alpha=0.2, color='blue', label='Estimate Range')
    ax1.plot(lot_year_groupby['lot_year'], lot_year_groupby['estimate_median'], 
             'b--', label='Estimate Median', alpha=0.7)
    ax1.plot(lot_year_groupby['lot_year'], lot_year_groupby['price_USD'], 
             'r-', label='Actual Price', linewidth=2)
    ax1.set_ylabel('Price (USD) - Log Scale')
    ax1.set_xlim(lot_year_groupby['lot_year'].min(), lot_year_groupby['lot_year'].max()+1)
    ax1.xaxis.set_major_locator(plt.MultipleLocator(1))  # 1년 단위로 눈금 설정
    ax1.legend(loc='upper left')
    
    # 두 번째 플롯: 거래량
    ax2.bar(lot_year_groupby['lot_year'], lot_year_groupby['count'], 
            color='gray', alpha=0.5, label='Count')
    ax2.set_ylabel('Count')
    ax2.set_xlim(lot_year_groupby['lot_year'].min(), lot_year_groupby['lot_year'].max()+1)
    ax2.xaxis.set_major_locator(plt.MultipleLocator(1))  # 1년 단위로 눈금 설정
    ax2.legend(loc='upper left')
    
    # 세 번째 플롯: 예상가 대비 실제가 차이 (퍼센트)와 이동평균
    bars = ax3.bar(lot_year_groupby['lot_year'], lot_year_groupby['price_diff_pct'], 
                   color=['red' if x < 0 else 'green' for x in lot_year_groupby['price_diff_pct']], 
                   alpha=0.5, label='Price Difference %')
    ax3.plot(lot_year_groupby['lot_year'], lot_year_groupby['ewm_diff'], 
             'b-', label='EWM (3-year span)', linewidth=2)
    
    # 다음 년도 예측 표시
    next_year = last_year + 1
    ax3.plot([last_year, next_year], [last_ewm, last_ewm], 'b--', alpha=0.5)
    ax3.fill_between([last_year, next_year], 
                    [prediction_lower, prediction_lower],
                    [prediction_upper, prediction_upper],
                    color='blue', alpha=0.2, label='Prediction Range')
    
    ax3.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    ax3.set_ylabel('Difference from Estimate (%)')
    ax3.set_xlim(lot_year_groupby['lot_year'].min(), next_year)
    ax3.set_xlabel('Year')
    ax3.xaxis.set_major_locator(plt.MultipleLocator(1))  # 1년 단위로 눈금 설정
    ax3.legend(loc='upper left', fontsize='7')
    
    # 예측 정보 표시
    prediction_text = f'Next Year ({next_year}) Prediction:\n{last_ewm:.1f}% ± {recent_std:.1f}%'
    ax3.text(0.98, 0.95, prediction_text, fontdict={'size': 8},
             transform=ax3.transAxes, bbox=dict(facecolor='white', alpha=0.8))
    
    plt.suptitle(f'Picasso {meta} Price Trend and Estimate Comparison with Prediction (filtered {filter_percent*100}%)')
    plt.tight_layout()
    plt.show()
    
    # 예측 결과 출력
    print(f"\n{meta} - 다음 년도({next_year}) 예측:")
    print(f"예상 차이: {last_ewm:.1f}%")
    print(f"예측 범위: {prediction_lower:.1f}% ~ {prediction_upper:.1f}%")
```

## 분야 Groupby 해서 비싼 평균가격 상위 3개

```python
# 평균 가격, 크기 및 작품 수 계산
type_stats = df_concat.groupby('type').agg({
    'price_USD': 'mean',
    'type': 'size' 
}).sort_values('price_USD', ascending=False)

type_stats.rename(columns={'type': 'Count', 'price_USD':'Average Price'}, inplace=True)
type_stats
```

```python
# Convert end_date to year and extract artwork type
df_concat['lot_year'] = pd.to_datetime(df_concat['end_date']).dt.year

result = pd.crosstab(df_concat['lot_year'], df_concat['type'])

print("\nArtwork Types by Year:")
result
```

```python
import matplotlib.pyplot as plt

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# 막대 그래프 생성 (두 개의 y축 사용)
fig, ax1 = plt.subplots(figsize=(12, 6))

# 첫 번째 y축 (평균 가격)
bars = ax1.bar(type_stats.index, type_stats['Average Price'], color='skyblue')
ax1.set_xlabel('Artwork Type')
ax1.set_ylabel('Average Price (USD)', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# x축 레이블 회전
plt.xticks(rotation=45, ha='right')

# 두 번째 y축 (작품 수)
ax2 = ax1.twinx()
line = ax2.plot(type_stats.index, type_stats['Count'], color='red', marker='o')
ax2.set_ylabel('Number of Artworks', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# 제목 설정
plt.title('Average Price and Count by Artwork Type')

# 레이아웃 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```
